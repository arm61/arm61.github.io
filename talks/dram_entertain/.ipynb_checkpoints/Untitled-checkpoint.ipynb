```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
np.random.seed(1)
```

```python
def straight_line(x, m, c):
    return m * x + c
```

```python
x = np.linspace(1, 10, 10)
```

```python
y = straight_line(x, 2., 1) + np.random.randn(10) * 0.6
```

```python
dy = 0.1 * y+ np.random.randn(10) * 0.6
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.errorbar(x, y, dy, marker='o', ls='', color='#4098d6')
ax.set_xlim(0, 11)
ax.set_ylim(0, 25)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_facecolor('white')
fig.savefig('xy.png')
```
