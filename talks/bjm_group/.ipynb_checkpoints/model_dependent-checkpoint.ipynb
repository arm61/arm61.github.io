```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import _fig_params
```

```python
np.random.seed(5)
```

```python
def arrhenius(x, ea, k):
    return k * np.exp(-ea / (8.617e-5 * x))
```

```python
T = np.array([-9.5, 0.5, 10.5, 19.5, 30, 50]) + 273.15
```

```python
D = np.exp(-0.62/(8.617e-5 * T)) * 9e11 + np.random.randn(T.size) * 0.5
Derr = np.ones_like(D) * 1
```

```python
plt.errorbar(T, D, Derr, marker='o', ls='')
plt.ylabel('k/s$^{-1}$')
plt.xlabel('1000T$^{-1}$/K$^{-1}$')
plt.savefig('figures/data.png', dpi=600)
plt.show()
```

```python
popt, pcov = curve_fit(arrhenius, T, D)
```

```python
xk = np.linspace(2.8, 4.2, 100)
```

```python
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
x = 1000/T
y = D
yerr = Derr
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
zstore = []
for i in range(len(y)):
    x1 = np.array([x[i]]*100)
    x2 = np.logspace(-4, 3, 100)
    x3 = norm.pdf(xk, loc=y[i], scale=yerr[i])
    x4 = norm.pdf(x2, loc=y[i], scale=yerr[i])
    #verts = [(x1[i],xk[i],x3[i]) for i in range(len(x1))] + [(x1.max(),0,0),(x1.min(),0,0)]
    #ax.add_collection3d(Poly3DCollection([verts],color=_fig_params.colors[0], alpha=0.5))
    ax.plot(x1, np.log(x2), x4, color=_fig_params.colors[0], alpha=1)
    zstore.append(norm.pdf(straight_line(x[i], *popt), loc=y[i], scale=yerr[i]))
#ax.plot(x, straight_line(x, *popt), zstore, '-', color=_fig_params.colors[1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
def log_likelihood(theta, x, y, yerr):
    m, b = theta
    model = m * x + b
    sigma2 = yerr ** 2
    return -0.5 * np.sum((y - model) ** 2 / sigma2 + np.log(sigma2))
```

```python
import emcee

pos = [1.25, 2.38] + 1e-4 * np.random.randn(32, 2)
nwalkers, ndim = pos.shape

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_likelihood, args=(x, y, yerr))
sampler.run_mcmc(pos, 5000, progress=True);
```

```python
flat_samples = sampler.get_chain(discard=100, thin=15, flat=True)
print(flat_samples.shape)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(122, projection='3d')
for i in range(len(y)):
    x1 = np.array([x[i]]*100)
    x2 = np.linspace(y[i] - yerr[i], y[i] + yerr[i], 100)
    x3 = norm.pdf(xk, loc=y[i], scale=yerr[i])
    x4 = norm.pdf(x2, loc=y[i], scale=yerr[i])
    verts = [(x1[i],xk[i],x3[i]) for i in range(len(x1))] + [(x1.max(),0,0),(x1.min(),0,0)]
    ax.add_collection3d(Poly3DCollection([verts],color=_fig_params.colors[0], alpha=0.5))
    ax.plot(x1, x2, x4, color=_fig_params.colors[0])
j = np.random.randint(0, flat_samples.shape[0], size=100)
for js in j:
    zstore = []
    for i in range(len(y)):
        zstore.append(norm.pdf(straight_line(x[i], *flat_samples[js]), loc=y[i], scale=yerr[i]))
    ax.plot(x, straight_line(x, *flat_samples[js]), zstore, '-', color=_fig_params.colors[1], alpha=0.1)
ax.set_zlim(0, 0.25)    
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.grid(True)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax = fig.add_subplot(121)
ax.errorbar(x, y, yerr, marker='o', ls='', zorder=10)
choice = np.random.randint(0, flat_samples.shape[0], size=100)
for i in j:
    ax.plot(x, straight_line(x, *flat_samples[i]), alpha=0.1, color=_fig_params.colors[1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.text(0.05, 0.95, '$m={0:.2f}\pm{2:.2f}$\n$c={1:.2f}\pm{3:.2f}$'.format(*flat_samples.mean(axis=0), *flat_samples.std(axis=0)), horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
#plt.savefig('figures/bayes_line.png', dpi=600)
```

We believe that $x$ and $y$ are linearly dependent, with some offset, so we *model* this with a straight line.

$$ y = m x + c $$

```python
def straight_line(x, m, c):
    return m * x + c
```

```python
popt, pcov = curve_fit(straight_line, x, y, sigma=yerr)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.errorbar(x, y, yerr, marker='o', ls='')
ax.plot(x, straight_line(x, *popt))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.text(0.05, 0.95, '$m={:.2f}\pm{:.2f}$\n$c={:.2f}\pm{:.2f}$'.format(
    popt[0], np.sqrt(pcov[0][0]), popt[1], np.sqrt(pcov[1][1])), horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
plt.savefig('figures/straight_line.png', dpi=600)
```

Where the values of the model parameters are.

```python
print('m: {:.2f} +/- {:.2f}'.format(popt[0], np.sqrt(pcov[0][0])))
print('c: {:.2f} +/- {:.2f}'.format(popt[1], np.sqrt(pcov[1][1])))
```

Ofcourse, the model can be a lot more complex.
