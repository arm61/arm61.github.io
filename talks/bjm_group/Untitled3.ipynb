```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import _fig_params
```

```python
n = norm(loc=10, scale=1)
x = np.linspace(5, 15, 1000)
plt.plot(x, n.pdf(x))
plt.text(5, 0.35, '$\mathcal{N}(10, 1^2)$')
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.tight_layout()
```
